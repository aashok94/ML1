```python
import numpy as np
from numpy import array as a
import matplotlib.pyplot as plot
from sklearn.datasets import make_moons
from scipy.spatial import distance_matrix as distMat

n_samples=300

#moons
x_moons, y_moons = make_moons(n_samples*2 , noise=0.05, shuffle = False)
x0 = x_moons[:n_samples]
x1 = x_moons[n_samples:]

#normalize using sample mean
n = (2* n_samples) - 1
x_mean = np.mean(x_moons, axis = 0)
new_xmoons = x_moons - x_mean
x_moons= np.matrix(x_moons)
print("Old x_moons \n", x_moons)
print("")

print("Mean =", x_mean)
print("")

new_xmoons= np.matrix(new_xmoons)
print("New x_moons \n", new_xmoons.shape)
print("")

scatter = new_xmoons.transpose() * new_xmoons
print("Scatter Matrix \n", scatter)
print("")
sample_coVar = scatter / n
print("Covarience Matrix \n", sample_coVar)

# for eigenvalue
w,v = np.linalg.eig(sample_coVar)
print("")
print("Eigen value \n", w)
print("")
print("Eigen Vector \n", v)
print("")

linear_pca = v[:,:2]
linear_pca = np.matrix(linear_pca)

x0_score = a(x0*linear_pca)
x1_score = a(x1*linear_pca)


x0_score1 = a(x0_score[:,0])
x0_score2 = a(x0_score[:,1]) 
x1_score1 = a(x1_score[:,0])
x1_score2 = a(x1_score[:,1])

plot.title("Linear PCA")
plot.scatter(x0_score1, x0_score2)
plot.scatter(x1_score1, x1_score2)
plot.show()


#kernel pca
sigma = 0.1
K = distMat(x_moons,x_moons)
K = K**2
K = -K*(1.0/(2*sigma**2))
K = np.exp(K)
# print("Kernel Matrix \n", K)
# print("")

x0_k = K[:int(x_moons.shape[0]/2),:]
x1_k = K[int(x_moons.shape[0]/2):,:]

wk,vk = np.linalg.eig(K)
K = np.matrix(K)

kernel_pca = vk[:,:2]
kernel_pca = np.matrix(kernel_pca)

x0_kscore = a(x0_k*kernel_pca)
x1_kscore = a(x1_k*kernel_pca)

x0_kscore1 = a(x0_kscore[:,0])
x0_kscore2 = a(x0_kscore[:,1]) 
x1_kscore1 = a(x1_kscore[:,0])
x1_kscore2 = a(x1_kscore[:,1])

plot.title("Kernel PCA")
plot.scatter(x0_kscore1, x0_kscore2)
plot.scatter(x1_kscore1, x1_kscore2)
plot.show()
plot.savefig('kpca.png')
```
